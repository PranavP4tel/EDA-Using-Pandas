```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r".\Excel files\Marathon.csv")
```

```python
print(df.head())
```

```python
#Filtering data for race length as 50km or 50mi, events held in USA and in the year 2020

df1  = df[(df["Event distance/length"].isin(["50km","50mi"])) & (df["Year of event"] == 2020) & (df["Event name"].str.split("(").str.get(1).str.split(")").str.get(0) == "USA") ]
```

```python
print(df1.head())
```

```python
df1.shape
```

```python
#Removing (USA) from event name
```

```python
df1["Event name"] = df1["Event name"].str.split("(").str.get(0)
```

```python
print(df1.head())
```

```python
#Creating up athlete age
```

```python
df1["Athlete Age"] = df1["Year of event"] - df1["Athlete year of birth"]
```

```python
print(df1.head())
```

```python
df2 = df1.copy()
```

```python
#Removing hours from athlete performance
df2["Athlete performance"] = df2["Athlete performance"].str.split(" ").str.get(0)
```

```python
df2.head()
```

```python
#Dropping columns: Athlete club, country, and year of birth, Athlete age category
df3 = df2.drop(["Athlete club","Athlete country","Athlete year of birth","Athlete age category"],axis = 1)
df3.head()
```

```python
df3.isna().sum()
```

```python
df3[df3["Athlete Age"].isna()]
```

```python
#dropping null values
df3 = df3.dropna()
```

```python
df3.shape
```

```python
#check for duplicates
df3[df3.duplicated()]
```

```python
#Reset index
df3.reset_index(drop=True)
```

```python
#Correcting data types
df3["Athlete Age"] = df3["Athlete Age"].astype(int)
```

```python
df3["Athlete average speed"] = df3["Athlete average speed"].astype(float) 
```

```python
df3.dtypes
```

```python
df3 = df3.rename(columns = {'Year of event':'year',
                            'Event dates':'date',
                            'Event name':'name',
                            'Event distance/length':'distance',
                            'Event number of finishers':'finishers',
                            'Athlete performance':'performance',
                            'Athlete gender':'gender',
                            'Athlete average speed':'avg_speed',
                            'Athlete ID':'id',
                            'Athlete Age':'age'
                           })
```

```python
df3.head()
```

```python
#reorder columns
df4 = df3[['date','name','distance','finishers','id','gender','age','performance','avg_speed','year']]
```

```python
df4.head()
```

```python
#Count of races as 50km or 50 mi
sns.histplot(df4['distance'])
```

```python
#Distribution of average speed of athletes for 50mi races
sns.histplot(df4[df4["distance"] == "50mi"]["avg_speed"])
```

```python
#Distribution of average speed of athleted for different race lengths, categorized by gender
sns.boxplot(df4,x = 'distance',y = 'avg_speed',hue = "gender")
```

```python
#Distribution of average speed of atheletes of different ages, categorized by gender
sns.lmplot(df4,x = 'age',y='avg_speed',hue = 'gender')
```

```python
#Difference in speed for males and females
```

```python
df4.groupby(['distance','gender'])['avg_speed'].mean()
```

```python
#age groups performing best in 50 m race (count 20 and above)
df4.query('distance == "50mi"').groupby('age')['avg_speed'].agg(['mean','count']).sort_values('mean',ascending = False).query('count>19')
```

```python
#Creating a column to indicate season in which event was held
df5 = df4.copy()
df5['month'] = df5['date'].str.split('.').str.get(1).astype(int)
```

```python
df5.head()
```

```python
df5['season'] = df5['month'].apply(lambda x:'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else "Winter")
```

```python
df5['season']
```

```python
df5['season'].nunique()
```

```python
#Calculating average speed of athletes categorized by seasons in which events were held
df5.groupby('season')['avg_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
